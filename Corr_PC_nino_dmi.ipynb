```python
from scipy import stats
import numpy as np
import xarray as xr
from scipy.stats import pearsonr
import pandas as pd
```

```python
ds1=xr.open_dataset('Corr_SSHanom_R1_92-17.nc')
ds2=xr.open_dataset('Corr_SSLanom_R1_92-17.nc')
ds3=xr.open_dataset('Corr_TSLanom_R1_92-17.nc')
ds4=xr.open_dataset('Corr_HSLanom_R1_92-17.nc')
```

```python
ds1
```

```python
ssh=ds1.SSHANOM
ssh1=ssh.to_dataframe()
ssh1.to_csv("sshanom_R1.csv")
```

```python
ssl=ds2.SSLANOM
ssl1=ssl.to_dataframe()
ssl1.to_csv("sslanom_R1.csv")
```

```python
tsl=ds3.TSLANOM
tsl1=tsl.to_dataframe()
tsl1.to_csv("tslanom_R1.csv")
```

```python
hsl=ds4.HSLANOM
hsl1=hsl.to_dataframe()
hsl1.to_csv("hslanom_R1.csv")
```

```python
ds4=pd.read_csv('sshanom_R1.csv')
ds5=pd.read_csv('sslanom_R1.csv')
ssh11=ds4["SSHANOM"]
ssl11=ds5["SSLANOM"]

stat, p= pearsonr(ssh11,ssl11)
print ("stat=%.3f,p=%.25f"%(stat,p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

```python
ds6=pd.read_csv('tslanom_R1.csv')
ds7=pd.read_csv('hslanom_R1.csv')
tsl11=ds6["TSLANOM"]
hsl11=ds7["HSLANOM"]

stat, p= pearsonr(ssl11,tsl11)
print ("stat=%.3f,p=%.25f"%(stat,p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

```python
stat, p= pearsonr(ssl11,hsl11)
print ("stat=%.3f,p=%.25f"%(stat,p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```
